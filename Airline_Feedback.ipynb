# Data Preprocessing

```python
df=pd.read_csv("/content/drive/My Drive/MINOR 6 SEM/airline.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
total_delay=df['Departure Delay in Minutes']+df ['Arrival Delay in Minutes']
df.insert(loc=9, column="Delay", value=total_delay)
```

```python
df.head()
```

```python
df1=df.dropna(subset=['Departure Delay in Minutes','Arrival Delay in Minutes'])
```

```python
df1.head()
```

```python
del df1['Arrival Delay in Minutes']
del df1['Departure Delay in Minutes']
```

```python
df1
```

```python
def tonum(input):
  if input=="neutral or dissatisfied":
    return 0
  else:
    return 1
```

```python
df1['satisfaction']=df1['satisfaction'].apply(tonum)
```

```python
def genderchange(input):
  if input=="male":
    return 0
  else : 
    return 1
```

```python
df1['Gender']=df1['Gender'].apply(genderchange)
```

```python
def bchange(input):
  if input=="Business":
    return 0
  elif input =="Eco Plus": 
    return 2
  else :
    return 1
```

```python
df1['Class']=df1['Class'].apply(bchange)
```

```python
del df1['Unnamed: 0']
del df1['id']
del df1['Customer Type']
del df1['Type of Travel']
```

# C4.5 Model Vislualization

```python
airline_c45_df=pd.read_csv("/content/drive/My Drive/MINOR 6 SEM/C4.5/C4.5_Airline_Prediction.csv")
```

```python
airline_c45_df.fillna(value="-", inplace=True)
airline_c45_df
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt

# Create the histogram using seaborn
sns.histplot(airline_c45_df['error'], kde=False)

# Get the counts of each bin
counts, bins, _ = plt.hist(airline_c45_df['error'], bins=20)
sns.set(style="darkgrid")


plt.text(0, 96969, "96947", ha='center', fontsize=12)
plt.text(1, 6699, "6647", ha='center', fontsize=12)

# Set the x and y axis labels
plt.xlabel('''Error
Accuracy = 93.584% ''')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
X = airline_c45_df['actual']
y = airline_c45_df['predicted']

cm = confusion_matrix(X, y)


# Plot the confusion matrix using Seaborn

plt.figure(figsize=(8,5))
sns.heatmap(cm , annot=True,fmt = 'd')
plt.xlabel("PREDICTION")
plt.ylabel("ACTUAL")
plt.show()
```

# Comparison of the Algorithms

```python
comparison_df=pd.read_csv("/content/drive/My Drive/MINOR 6 SEM/comparison.csv")
comparison_df
```

```python
comparison_df = comparison_df.drop(['Recall', 'MCC', 'ROC Area', 'PRC Area'], axis=1)
sns.set_style("darkgrid")

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(7, 8))

for i, metric in enumerate(['TP Rate', 'FP Rate', 'Precision',  'F-Measure']):
    sns.barplot(x='Algorithm', y=metric, data=comparison_df, ax=axs[i//2, i%2])
    axs[i//2, i%2].set_title(metric)
    axs[i//2, i%2].set_xlabel('Algorithm')
    axs[i//2, i%2].set_ylabel(metric)
    for p in axs[i//2, i%2].patches:
        axs[i//2, i%2].annotate(f"{p.get_height():.3f}", 
                                 (p.get_x() + p.get_width() / 2., p.get_height()/2), 
                                 ha='center', va='center', 
                                 fontsize=14, color='white', xytext=(0, 5),
                                 textcoords='offset points', weight='bold')

plt.tight_layout()

plt.show()
```
